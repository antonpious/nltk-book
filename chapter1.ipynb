```python
# TODO multiple words by the api is taking words together
# no option to send an OR command between words
print("Displaying word 'nation' occurrences")
text4.concordance("nation")
print("Displaying word 'terror' occurrences")
text4.concordance("terror")
print("Displaying word 'god' occurrences")
text4.concordance("god")
```

```python
#if you want to look at the occurrences above 25 count
# TODO find out how can this be one call
# Currently you need to use concordance get the count
# Then come back here and get the line list
# NOTE python style guide uses snake case instead of camelCase in 
# many programming languages
# link to google style guide
# https://google.github.io/styleguide/pyguide.html#316-naming
nation_list = text4.concordance_list("nation", lines=348)
print("Total occurrence: ", len(nation_list))
# 0 based so printing the 26th occurrence
occurrence_line_no = 25
occurrence_line_output = nation_list[occurrence_line_no].line
print(f"{occurrence_line_no+1 }th Occurrence:{occurrence_line_output}")
```
